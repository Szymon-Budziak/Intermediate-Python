# Lambda

```python
add10 = lambda x: x+10
print(add10(5))
```

```python
def add10_func(x):
    return x+10
# this is the same but lambda is shorter and is only in one line
```

```python
mult = lambda x,y: x*y
print(mult(2,7))
```

### SORTED

```python
points2D = [(1,2),(17,1),(-4,-11),(18,54)]
points2D_sorted = sorted(points2D)
print(points2D)
print(points2D_sorted) # by default it sorts our list by the first argument
```

```python
points2D_sorted2 = sorted(points2D, key=lambda x: x[1]) # it sorts our list by the second argument
print(points2D_sorted2) # we can also crerate a def and put it in tke key instead of lambda but lambda is shorter
```

```python
points2D_sorted3 = sorted(points2D, key = lambda x: x[0] + x[1])
print(points2D_sorted3) # it is sorted according to a sum of each tuple (-4-11=-15, 1+2=3... : 15<3<...)
```

### MAP

```python
# map(function, sequence)
a = [1,2,3,4,5]
b = map(lambda x: x*2, a)
print(b)
print(list(b))
```

```python
# we can get the same thing with list comprehension
c = [x*2 for x in a]
print(c)
```

```python
# filter(function, sequence) function must return True or False and filter return all True elements
a = [1,2,3,4,5,6]
b = filter(lambda x: x%2==0,a) # only even numbers
print(b)
print(list(b))
```

```python
# also we can get the same thing with list comprehension
c = [x for x in a if x%2==0]
print(c)
```

### REDUCE

```python
# reduce(function, sequence)
from functools import reduce
a = [1,2,3,4,5]
product_a = reduce(lambda x,y: x*y,a)
print(product_a) # 1*2=2*3=6*4=24*5=120
```
