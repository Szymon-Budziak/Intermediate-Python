### String formatting

```python
variable1 = 'Szymon'
my_string11 = 'the variable is %s' % variable1
print(my_string11)
```

```python
variable2 = 5
my_string12 = 'the variable is %d' % variable2  # decimal value
print(my_string12)
```

```python
variable3 = 12.12345678
my_string13 = 'the variable is %.2f' % variable3
# .2f give us 2 digits of decimal point by default it has 6 digits after dot
print(my_string13)
```

```python
variable4 = 5.764
my_string14 = 'the variable is {}'.format(variable4)
print(my_string14)
```

```python
my_string15 = 'the variable is {:.2f}'.format(variable4)
# 2 digits after the decimal point
print(my_string15)
```

```python
variable5 = 67
my_string16 = 'the variable is {:.2f} and {}'.format(variable4, variable5)
print(my_string16)
```

```python
my_string17 = f"the variable is {variable4*2} and {variable5}"
print(my_string17)
```
