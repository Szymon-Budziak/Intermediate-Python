# Dictionaries

```python
mydict = {"name": "Max", "age": 28, "city": "New York"}
print(mydict)
```

```python
mydict2 = dict(name="Mary", age=19, city="Boston")
print(mydict2)
```

```python
value = mydict["name"]
value2 = mydict["age"]
print(value)
print(value2)
```

```python
# add a new key and value to the dictionary
mydict["email"] = "max@xyz.com"
print(mydict)
```

```python
# del deletes given key
del mydict["name"]
print(mydict)
```

```python
# pop removes given key and value
mydict.pop("age")
print(mydict)
```

```python
# popitem removes last item
mydict.popitem()
print(mydict)
```

```python
# check if there is an element in dictionary
try:
    print(mydict["lastname"])
except:
    print("Error")
```

```python
# loop through the dictionary
for key, value in mydict.items():
    print(key, value)
```

```python
# copy a dictionary
mydict_copy1 = mydict.copy()
mydict_copy2 = dict(mydict)
```

```python
# merge two dictionaries
mydict2 = {"name": "Max", "age": 29, "email": "max@xyz.com"}
mydict3 = dict(name="Mary", age=25, city="Boston")
mydict2.update(mydict3)
print(mydict2)
# we changed all existing keys and values and added new keys and values
```

```python
mydict4 = {3: 9, 6: 36, 9: 81}
print(mydict4)
value = mydict4[3]
print(value)
```

```python
# tuple as a key
mytuple = (8, 7)
mylist2 = [1, 3]
mydict5 = {mytuple: 16}
print(mydict5)
# if we use a list instead of tuple it is not possible as a key because list is muttable
```
