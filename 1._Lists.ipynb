# Lists

```python
myList = ["banana", "cherry", "apple"]
print(myList)
```

```python
myList2 = [5, True, "apple", "apple"]
print(myList2)
```

```python
# print element with index 0
item1 = myList[0]
print(item1)
```

```python
# print second element from the end
item2 = myList[-2]
print(item2)
```

```python
# go through the list with for loop
for i in myList:
    print(i)
```

```python
# insert a new element to the list with index
myList.insert(2, "bluberry")
print(myList)
```

```python
# pop element (remove it) and print it
item3 = myList.pop()
print(item3)
print(myList)
```

```python
# sort a list
new_list = sorted(myList)
print(myList)
print(new_list)
```

```python
# list of 5 zeros
myList3 = [0]*5
print(myList3)
```

```python
# concatenate lists
myList4 = [1, 2, 3, 4, 5]
new_list2 = myList3 + myList4
print(new_list2)
```

```python
# reverse a list with slicing and -1
myList5 = [1, 2, 3, 4, 5, 6, 7, 8, 9]
a = myList5[::-1]
print(a)
```

```python
# copy a list 1
list_original = ["banan", "cherry", "apple"]
list_copy = list_original.copy()
list_copy.append("lemon")
print(list_copy)
print(list_original)
```

```python
# copy a list 2
list_original = ["banan", "cherry", "apple"]
list_copy = list(list_original)
list_copy.append("lemon")
print(list_copy)
print(list_original)
```

```python
# copy a list 3
list_original = ["banan", "cherry", "apple"]
list_copy = list_original[:]
list_copy.append("lemon")
print(list_copy)
print(list_original)
```

```python
# list comprehension
myList6 = [1, 2, 3, 4, 5, 6]
c = [number*number for number in myList6]
print(myList6)
print(c)
```
