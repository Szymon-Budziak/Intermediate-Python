# collections: Counter. namedtuple, OrderedDict, defaultdict, deque

```python
from collections import Counter
```

```python
string = "aaaaaabbbbbbbbccccccc"
```

```python
my_counter = Counter(string) # we can use here a list or any other structure
print(my_counter)
```

```python
print(my_counter.items())
```

```python
my_counter.keys()
```

```python
my_counter.values()
```

```python
my_counter.most_common(1)
```

```python
my_counter.most_common(2)
```

```python
print(my_counter.most_common(1)[0])
```

```python
print(my_counter.most_common(1)[0][0])
```

```python
my_counter.elements()
```

```python
print(list(my_counter.elements()))
```

```python
from collections import namedtuple
```

```python
Point = namedtuple('Point', 'x,y') # this will create class Point with x and y fields
```

```python
pt = Point(1,-4)
```

```python
print(pt)
```

```python
print(pt.x, pt.y)
```

```python
from collections import OrderedDict
```

```python
# ordered dictionary remember the order when their items were inserted
```

```python
ordered_dict = OrderedDict()
```

```python
ordered_dict["c"] = 3
ordered_dict["d"] = 4
ordered_dict["b"] = 2
ordered_dict["e"] = 5
ordered_dict["a"] = 1
```

```python
print(ordered_dict)
```

```python
# with normal dictionary it also works since Python 3.7
```

```python
from collections import defaultdict
```

```python
d = defaultdict(int) #int here is a default type
```

```python
d['a'] = 1
d['b'] = 2
print(d)
```

```python
print(d['a'])
print(d['b'])
```

```python
print(d['c']) # c doesn't exist so it will return 0 which is default value of int (float -> 0.0)
```

```python
d2 = defaultdict(list)
```

```python
print(d2['c']) # c doesn't exist in d2 so it will print empty list because it is default value of a list
```

```python
from collections import deque
```

```python
de = deque()
```

```python
de.append(1)
de.append(2)
de.append(9)
```

```python
print(de)
```

```python
de.appendleft(7)
print(de)
```

```python
de.pop()
print(de)
```

```python
de.popleft()
print(de)
```

```python
de.clear()
print(de)
```

```python
de.extend([4,5,234])
print(de)
```

```python
de.extendleft([78,6,8]) # 8 will be first then 6 and 78
print(de)
```

```python
de.rotate(1) # it rotates all elements by 1 place to the right
print(de)
```

```python
de.rotate(-2) # it rotates all elements by 2 places to the left
print(de)
```
