```python
class ValueTooHighError(Exception): # base class is Exception class
    pass

def test_value(x):
    if x > 100:
        raise ValueTooHighError('Value is too high!')
        
try:
    test_value(200)
except ValueTooHighError as e:
    print(e)
```

```python
class ValueTooHighError(Exception): # base class is Exception class
    pass

class ValueToSmallError(Exception): # base class is Exception class
    def __init__(self, message, value):
        self.message = message
        self.value = value

def test_value(x):
    if x > 100:
        raise ValueTooHighError('Value is too high!')
    if x < 10:
        raise ValueToSmallError('Value is too small!',x)
        
try:
    test_value(1)
except ValueTooHighError as e:
    print(e)
except ValueToSmallError as e:
    print(e.message, e.value)
```
