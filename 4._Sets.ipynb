# Sets

```python
myset = {1, 2, 3, 4, 5, 1, 2, 3}
print(myset)
```

```python
myset1 = set("Hello")
print(myset1)
```

```python
# create new empty set
myset3 = set()
```

```python
# add element in a set
myset.add(8)
myset.add(65)
myset.add(54)
print(myset)
```

```python
# remove element in a set
myset.remove(5)
print(myset)
```

```python
# discard element in a set
myset.discard(434)
print(myset)
```

```python
# pop element in a set
print(myset.pop())
```

```python
# clear set
myset.clear()
print(myset)
```

```python
# combine elements from 2 sets without duplications using union
odds = {1, 3, 5, 7, 9}
evens = {0, 2, 4, 6, 8}
primes = {2, 3, 5, 7}
u = odds.union(evens)
print(u)
```

```python
# the same elements from 2 sets using intersection
i = odds.intersection(primes)
print(i)
```

```python
# calculate the difference between two sets
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
# set without elements that are both in set A and set B using difference
diff = setA.difference(setB)
print(diff)
```

```python
# set with all elements from set A and set B but without elements that are in both sets using symmetric_difference
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
diff_sym = setA.symmetric_difference(setB)
print(diff_sym)
```

```python
# update the set by adding elements that are in the second set using update
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
setA.update(setB)
print(setA)
```

```python
# update the set by keeping only elements that are in both sets using intersection_update
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
setA.intersection_update(setB)
print(setA)
```

```python
# update the set by removing elements from set that are also in the second set using difference_update
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
setA.difference_update(setB)
print(setA)
```

```python
# update a set by keeping only elements found in set A and set B but without elements that are in both sets using symmetric_difference_update
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
setA.symmetric_difference_update(setB)
print(setA)
```

```python
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
setC = {1, 2, 3, 4, 5, 6}
setD = {1, 2, 3}
setE = {6754, 654, 61}
# is a set a subset of another set
print(setC.issubset(setD))
print(setD.issubset(setC))  # 123 are also in 123456
# set D is a subset of set C
```

```python
# is a set a superset of another set
print(setC.issuperset(setD))
print(setD.issuperset(setC)) # 123456 contains 123
# set C is a superset of set D
```

```python
# do sets have the same elements or not
print(setA.isdisjoint(setB))  # False because they have same elements
print(setA.isdisjoint(setE))  # True because they don't have same elements
```

```python
# copying sets
setC_copy = setC.copy()
setD_copy = set(setD)
print(setC_copy)
print(setD_copy)
```
