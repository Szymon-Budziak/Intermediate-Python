# Tuples

```python
mytuple1 = ("Max", 28, "Boston")
print(mytuple1)
```

```python
# without parenthesis it is still tuple
mytuple1 = "Max", 28, "Boston"
print(mytuple1)
```

```python
# this is not a tuple
not_a_tuple = ("Max")
print(type(not_a_tuple))
```

```python
# we have to add comma to change it to a tuple
this_is_tuple = ("Max",)
print(type(this_is_tuple))
```

```python
# list converted to a tuple
mytuple2 = tuple(["Max", 28, "Boston"])
print(mytuple2)
```

```python
item = mytuple2[-2]
print(item)
```

```python
for i in mytuple2:
    print(i)
```

```python
# len of tuple
mytuple3 = ('a', 'p', 'p', 'l', 'e')
print(len(mytuple3))
```

```python
# count how many "p" is in tuple
print(mytuple3.count('p'))
```

```python
# first index of specified element in tuple
print(mytuple3.index('p'))
```

```python
# convert tuple into a list
myList = list(mytuple3)
print(myList)
```

```python
# convert list into a tuple
mytuple4 = tuple(myList)
print(mytuple4)
```

```python
a = (1, 2, 3, 4, 5, 6, 7, 8, 9)
b = a[6:2:-1]
print(b)
```

```python
mytuple5 = "Max", 28, "Boston"
name, age, city = mytuple5
print(name)
print(age)
print(city)
```

```python
# all elements between i1 and i3 converted in a list
mytuple6 = (0, 1, 2, 3, 4, 5, 6, 7)
i1, *i2, i3 = mytuple6
print(i1)
print(i3)
print(i2)
```
