# itertools: product, permutations, combinations, accumulate, groupby and infinite iterators

```python
from itertools import product
```

```python
a = [1,2]
b = [3,4]
prod = product(a,b)
print(list(prod)) # kartezian product
```

```python
c = [5]
prod2 = product(a,c,repeat=2)
print(list(prod2))
```

```python
from itertools import permutations
```

```python
a = [1,2,3]
perm = permutations(a)
print(list(perm))
```

```python
perm2 = permutations(a,2) # 2 is a length of the permutation
print(list(perm2))
```

```python
from itertools import combinations, combinations_with_replacement
```

```python
a = [1,2,3,4]
comb = combinations(a,2) # 2 is a length of the combination
print(list(comb)) # there are no repetitions
```

```python
comb_wr = combinations_with_replacement(a,2) # 2 is a length of the combinations with replacement
print(list(comb_wr)) # here are repetitions
```

```python
from itertools import accumulate
import operator
```

```python
a = [1,2,3,4]
acc = accumulate(a) # by default, accumulate compute the sums
print(a)
print(list(acc)) # 1 | 1+2=3 | 3+3=6 | 6+4=10
```

```python
acc2 = accumulate(a, func=operator.mul)
print(list(acc2)) # 1 | 1*2=2 | 2*3=6 | 6*4=24
```

```python
c = [1,2,5,3,4,6]
print(c)
acc3 = accumulate(c, func=max)
print(list(acc3))
```

```python
from itertools import groupby
```

```python
a = [1,2,3,4,5]
def smaller_than_3(x):
    return x < 3
group_obj = groupby(a, key=smaller_than_3)
print(group_obj)
```

```python
a = [1,2,3,4,5]
def smaller_than_3(x):
    return x < 3
group_obj = groupby(a, key=smaller_than_3)
for key, value in group_obj:
    print(key,list(value))# key is True or False, because number is <3 or >3
```

```python
# the same with lambda expression
group_obj = groupby(a, key=lambda x: x<3)
for key, value in group_obj:
    print(key,list(value))
```

```python
persons =[{'name': 'Szymon', 'age': 19},{'name': 'Janusz', 'age': 19},
          {'name': 'Weronika', 'age': 18},{'name': 'Luiziana', 'age': 23}]
persons_obj = groupby(persons, key=lambda x: x['age'])
for key, value in persons_obj:
    print(key,list(value))
```

```python
from itertools import count, cycle, repeat
```

```python
for i in count(10):
    print(i) # infinite loop that starts at 10, ads 1 for every repetition
    if i == 15:
        break
```

```python
a = [1,2,3,4]
for i in cycle(a):
    print(i) # infinite loop through items from iterator (list)
```

```python
for i in repeat(1, 4): # it repeats 1 for 4 times (4 is a second argument which is stop-how many times i want to repeat it)
    print(i)
```
